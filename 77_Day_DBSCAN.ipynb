```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
X=np.array([[1,2],[2,2],[3,4],[2,8],[8,8],[25,88]])
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.cluster import DBSCAN

db=DBSCAN(eps=10,min_samples=3)
cluster=db.fit_predict(X)
db.fit(X)
db.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=cluster, cmap='viridis')
```

```python
from sklearn.datasets import make_circles

X=make_circles(n_samples=500,factor=0.1,noise=0.1, random_state=22)
plt.scatter(X[0][:,0],X[0][:,1])
```

```python
db=DBSCAN(eps=0.3,min_samples=6)

clusters = db.fit_predict(X[0])

plt.scatter(X[0][:,0],X[0][:,1],c=clusters, cmap='viridis')
```
