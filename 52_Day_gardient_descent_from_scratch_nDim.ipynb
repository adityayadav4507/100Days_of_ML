```python
from sklearn.datasets import load_diabetes
import numpy as np
import pandas as pd
```

```python
X,y = load_diabetes(return_X_y=True,as_frame=True)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred = lr.predict(X_test)
r2 = r2_score(y_test,y_pred)
print(r2)
```

```python
from sklearn.model_selection import cross_val_score
cross = np.mean(cross_val_score(lr,X,y,scoring="r2",cv=5))
cross
```

# make our own class

```python
class GDreg:

  def __init__(self,learning_rate=0.01,epochs=100):
    self.lr = learning_rate
    self.epochs = epochs
    self.coef_ = None
    self.intercept_= None

  def fit(self,X_train,y_train):
    # initilization
    self.intercept_ = 0
    self.coef_ = np.zeros(X_train.shape[1])

    for i in range(self.epochs):

      hat=self.intercept_ + np.dot(X_train,self.coef_)
     # print("hat=",hat.shape)
      der=-2*np.mean(y_train-hat)
     # print("der=",der.shape)
      self.intercept_-= (self.lr *der)

     # print("err=",(y_train-hat).shape)

      der1=np.dot((y_train-hat).T,X_train) # it store np.sum(X(y-y_pre)) X-> crosponding column of beta
                                           # ex- np.sum(Xi1(yi-y_pred)). beta_1 ke crosponging

      # print(der1.shape)

      # convert der1 into derivative
      der_convert=(-2/len(X_train))*der1  # der1/len(X_train) = mean  , len(X_train) =n
      # print(der_convert.shape)
      self.coef_ -= (self.lr * der_convert)

    print(self.intercept_,self.coef_)

  def predict(self,X_test):
    return np.dot(X_test,self.coef_) + self.intercept_
```
