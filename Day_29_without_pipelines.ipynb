```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('/content/titanic.csv')
```

```python
df.head()
```

```python
# unnecessary column remove
df.drop(columns=['PassengerId','Name','Ticket','Cabin'] ,inplace=True)
```

```python
df.head()
```

# Step -1 -> Train test split

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']),
                                                 df['Survived'],
                                                 test_size=0.2, # 20 data used in test
                                                random_state=42) # same spilt get when we run this code again and again
```

```python
X_train.head()
```

```python
df.isnull().sum()
```

```python
# Apply imputer
si_age = SimpleImputer() # by default mean mean is the strategy
si_embarked = SimpleImputer(strategy='most_frequent')

X_train_age = si_age.fit_transform(X_train[['Age']]) # this is new array original df no changes
X_train_embarked = si_embarked.fit_transform(X_train[['Embarked']])

X_test_age = si_age.transform(X_test[['Age']])
X_test_embarked = si_embarked.transform(X_test[['Embarked']])
```

```python
# Apply One Hot Encoding  on Sex and embarked  nominal category

ohe_sex = OneHotEncoder(sparse_output=False,handle_unknown='ignore') # two different object of OHE bec of one have missing vale (embarked)
ohe_embarked = OneHotEncoder(sparse_output=False,handle_unknown='ignore')

X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked_ohe = ohe_embarked.fit_transform(X_train_embarked) # not using original column bec of missing value in original column

X_test_sex = ohe_sex.transform(X_test[['Sex']])
X_test_embarked_ohe = ohe_embarked.transform(X_test_embarked)

```

```python
X_test.head(2)
```

```python
# we have 3 new array
# make new array train rem
X_train_rem= X_train.drop(columns=['Sex','Age','Embarked'])
X_test_rem= X_test.drop(columns=['Sex','Age','Embarked'])

```

```python
# combine new array (3) and x train rem column

X_train_transformed = np.concatenate((X_train_rem,X_train_age,X_train_embarked_ohe,X_train_sex), axis=1)
X_test_transformed = np.concatenate((X_test_rem,X_test_age,X_test_embarked_ohe,X_test_sex), axis=1)
```

```python
X_test_transformed.shape
```

```python
# train model
clf= DecisionTreeClassifier()
clf.fit(X_train_transformed,y_train)
```

```python
y_pred=clf.predict(X_test_transformed)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
import pickle # to deploy our model on server web dev part

import os # for creating directory
os.makedirs('model',exist_ok=True)
```

```python
pickle.dump(ohe_sex,open('model/ohe.sex.pkl' , 'wb'))
pickle.dump(ohe_embarked,open('model/ohe.embarked.pkl' , 'wb'))
pickle.dump(clf,open('model/clf.pkl','wb'))
```
