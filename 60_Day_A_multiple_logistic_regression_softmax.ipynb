```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/Iris.csv')
```

```python
df.head()
```

```python
df=df[['SepalLengthCm','SepalWidthCm','Species']]
```

```python
X=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
from sklearn.preprocessing import LabelEncoder

en=LabelEncoder()
y=en.fit_transform(y)
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2 , random_state=22)
clf=LogisticRegression(multi_class='multinomial')
clf.fit(X_train,y_train)
```

```python
clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , r2_score,confusion_matrix
```

```python
y_pred=clf.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

```python
pd.DataFrame(confusion_matrix(y_test,y_pred))
```

```python
# predict probability
clf.predict_proba(np.array([[2.22,4.22]])) # it give the probability of each class in which this test data belong
```

```python
clf.predict(np.array([[2.22,4.22]]))
```

```python
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt

plot_decision_regions(X.values ,y, clf ,legend=2)

plt.xlabel("x")
plt.show()
```
