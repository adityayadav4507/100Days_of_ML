```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/student_clustering.csv')
df
```

```python
plt.scatter(df['cgpa'],df['iq'],cmap='viridis')
plt.xlabel('cgpa')
plt.ylabel('iq')
plt.show()
```

```python
from sklearn.cluster import KMeans

wcss=[]

for i in range(1,10):
  km=KMeans(n_clusters=i,random_state=0)
  km.fit(df[['cgpa','iq']])
  wcss.append(km.inertia_)

wcss
```

```python
plt.plot(range(1,10),wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
X=df.iloc[:,:].values
km=KMeans(n_clusters=3,random_state=0)
y_predicted=km.fit_predict(X)
y_predicted
```

```python
X[y_predicted==0] # point which classified in 0 cluster
```

```python
X[y_predicted==0,0] # 0th cluster point of x axis
```

```python
plt.scatter(X[y_predicted==0,0],X[y_predicted==0,1],c='red')
plt.scatter(X[y_predicted==1,0],X[y_predicted==1,1],c='blue')
plt.scatter(X[y_predicted==2,0],X[y_predicted==2,1],c='green')
plt.scatter(X[y_predicted==3,0],X[y_predicted==3,1],c='yellow')
```

# K-Means in 3-D

```python
from sklearn.datasets import make_blobs

centroids = [(-5,-5,5),(5,2,3),(-2,-3,-5),(1,3,-3)]
cluster_std = [1,1,1,1]

X,y = make_blobs(n_samples=1000,centers=centroids,cluster_std=cluster_std,random_state=0)
```

```python
import plotly.express as px
fig=px.scatter_3d(x=X[:,0],y=X[:,1],z=X[:,2])
fig.show()
```

```python
from sklearn.cluster import KMeans

wcss=[]

for i in range(1,10):
  km=KMeans(n_clusters=i,random_state=0)
  km.fit(X)
  wcss.append(km.inertia_)

wcss
```

```python
plt.plot(range(1,10),wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
km=KMeans(n_clusters=4,random_state=0)
y_predicted=km.fit_predict(X)
y_predicted
```

```python
df=pd.DataFrame()

df['X']=X[:,0]
df['y']=X[:,1]
df['z']=X[:,2]
df['label']=y_predicted


fig=px.scatter_3d(df,x='X',y='y',z='z',color='label')
fig.show()
```
