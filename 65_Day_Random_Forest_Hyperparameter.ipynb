````python
import numpy as np
import pandas as pd
````

````python
df=pd.read_csv('/content/heart.csv')
````

````python
df.head()
````

````python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
````

````python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
````

````python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
clf.fit(X_train,y_train)
````

````python
from sklearn.ensemble import GradientBoostingClassifier
gb=GradientBoostingClassifier()
gb.fit(X_train,y_train)
````

````python
from sklearn.svm import SVC
svm=SVC()
svm.fit(X_train,y_train)
````

````python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
````

````python
from sklearn.metrics import accuracy_score
lr.predict(X_test)
print("lr= ",accuracy_score(y_test,lr.predict(X_test)))
svm.predict(X_test)
print("svm= ",accuracy_score(y_test,svm.predict(X_test)))
gb.predict(X_test)
print("gb= ",accuracy_score(y_test,gb.predict(X_test)))
clf.predict(X_test)
print("rf= ",accuracy_score(y_test,clf.predict(X_test)))
````

````python
from sklearn.model_selection import cross_val_score
print("lr= ",np.mean(cross_val_score(lr,X_train,y_train,cv=10,scoring='accuracy')))
print("svm= ",np.mean(cross_val_score(svm,X_train,y_train,cv=10,scoring='accuracy')))
print("gb= ",np.mean(cross_val_score(gb,X_train,y_train,cv=10,scoring='accuracy')))
print("rf= ",np.mean(cross_val_score(clf,X_train,y_train,cv=10,scoring='accuracy')))
````

````python
print("lr= ",np.mean(cross_val_score(LogisticRegression(max_iter=10000),X_train,y_train,cv=10,scoring='accuracy')))
````

````python

n_estimators=[20,60,100,120]
max_features=[0.3,0.5,1.0]

max_depth=[2,5,None]

max_samples=[0.5,0.75,1.0]

bootstrap=[True,False]

min_samples_split = [2,4,5]

min_samples_leaf = [1,2,3]
````

````python
params_grid={'n_estimators' : n_estimators,
           'max_features': max_features,
           'max_depth' : max_depth,
           'max_samples' : max_samples,
           'bootstrap' : bootstrap,
           'min_samples_split' : min_samples_split,
           'min_samples_leaf' : min_samples_leaf
           }

print(params_grid)
````

````python
from sklearn.model_selection import RandomizedSearchCV
rf=RandomForestClassifier()

Rand_cv=RandomizedSearchCV(estimator=rf,
                           param_distributions=params_grid,
                           cv=10,
                           n_jobs=-1,
                           verbose=2)
````

````python
Rand_cv.fit(X_train,y_train)
````

````python
Rand_cv.best_params_
````
