```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/weight-height.csv')
```

```python
df=df.drop('Gender',axis=1) # bec we train simple LR
```

```python
df.sample(4)
```

```python
plt.scatter(df['Height'],df['Weight'])
plt.xlabel('Height')
plt.ylabel('Weight')
# sort of liner data
```

```python
X=df.iloc[:,0:1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

# show the best fit line

```python
plt.scatter(df['Height'],df['Weight'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('Height')
plt.ylabel('Weight')
```

```python
# 1. slope of the line
m=lr.coef_
```

```python
# 2. intercept of the line
b= lr.intercept_
```

```python
# 3. line y=m*x + b
m*61.068054 + b
```

```python
# 4.straight line (no boundary)
m*61000 + b
```
