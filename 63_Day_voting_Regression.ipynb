```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
df=housing.data
df_target=housing.target
```

```python
df=pd.DataFrame(df,columns=housing.feature_names)
```

```python
df.sample(3)
```

```python
df['target']=df_target
```

```python
df.sample(3)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(df.drop('target',axis=1),df['target'],test_size=0.2,random_state=42)
```

```python
X_train.sample(3)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
lr=LinearRegression()
dt=DecisionTreeRegressor()
svr=SVR()
```

```python
estimators = [('lr',lr),('dt',dt),('svr',svr)]
```

```python
for estimator in estimators:
  scores=cross_val_score(estimator[1],X_train,y_train,scoring='r2',cv=10)
  print(estimator[0],np.round(np.mean(scores),2))
```

```python
from sklearn.ensemble import VotingRegressor
vr=VotingRegressor(estimators)
```

```python
scores = cross_val_score(vr,X_train,y_train,scoring='r2',cv=10)
print(np.round(np.mean(scores),2))
```
