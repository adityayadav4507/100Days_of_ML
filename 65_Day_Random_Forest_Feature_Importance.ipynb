````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
df=pd.read_csv('/content/train.csv')
````

````python
df.head()
````

````python
X=df.iloc[:,1:]
y=df.iloc[:,0]
````

````python
X.head()
````

````python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
````

````python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
classifier.fit(X_train,y_train)
````

````python
classifier.feature_importances_
````

````python
sns.heatmap(classifier.feature_importances_.reshape(28,28))
````

# calculate feature importance for decision tree


````python
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
# define the model
model = DecisionTreeClassifier()
# fit the model

X,y = make_classification(n_samples=5,n_classes=2, n_features=2, n_informative=2, n_redundant=0, random_state=1)
model.fit(X, y)
````

````python
from sklearn.tree import plot_tree
plot_tree(model)
````

````python
model.feature_importances_
````

````python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=3)
rf.fit(X,y)
````

````python
rf.feature_importances_
````

````python
imp1=rf.estimators_[0].feature_importances_
````

````python
imp2=rf.estimators_[1].feature_importances_
````

````python
imp3=rf.estimators_[2].feature_importances_
````

````python
print(imp1)
print(imp2)
print(imp3)
````

````python
print((imp1+imp2+imp3)/3) # random forest take mean
````
